```python
import re
from bs4 import BeautifulSoup 
def review_to_wordlist(review):
    '''
    Meant for converting each of the IMDB reviews into a list of words.
    '''
    # First remove the HTML.
    review_text = BeautifulSoup(review,"lxml").get_text()
    
    # Use regular expressions to only include words.
    review_text = re.sub("[^a-zA-Z]"," ", review_text)
    
    # Convert words to lower case and split them into separate words.
    words = review_text.lower().split()
   
    # Return a list of words
    return(words)
```

```python
%%time
traindata = []
train_words = []
for i in range(0,len(train['review'])):
    traindata.append(" ".join(review_to_wordlist(train['review'][i])))
    train_words.append(review_to_wordlist(train['review'][i]))
    
testdata = []
test_words = []
for i in range(0,len(test['review'])):
    testdata.append(" ".join(review_to_wordlist(test['review'][i])))
    test_words.append(review_to_wordlist(test['review'][i]))
```

#### 训练好词向量后一些常用的使用方法

#### 有了词向量后,需要把文本转换成特征矩阵.  
假如一个句子有50个词,经过word2vec以后,每个词转变为一个100维的向量.  直接替换的话,那每个句子就变成了5000个特征的特征矩阵,显然不能则么搞.
我们采用取均值的方法,可以这么理解：我们有一个N维空间,每个词就是N维空间里的一个点(或者说向量).一个句子有50个词,也就是说这个句子由N维空间里的50个点组成,现在我们想用N维空间中的某一个点
来表示这个句子,则我们把这N个点的向量加起来,(向量的加法,各个维度相加)，再取平均.这样我们就用N维空间里的一个点把这个句子表达了出来.


```python
X_all_new = []
for sent in X_all:
    X_all_new.append(np.mean([model.wv[w] for w in sent if w in model.wv],axis=0))
```

```python
print(len(X_all_new))
print(len(X_all_new[0]))
```

```python
lentrain = len(traindata)
X_train = X_all_new[:lentrain]  
X_test = X_all_new[lentrain:]

y_train = train['sentiment']
```

```python
print(len(X_train))
print(len(y_train))
```

### 上算法

```python
%%time
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
predict_result = log_reg.predict(X_test)
result_df = pd.DataFrame(data={"id":test["id"], "sentiment":predict_result})
result_df.to_csv("log_reg_word2vec.csv",index=False)
    
```
